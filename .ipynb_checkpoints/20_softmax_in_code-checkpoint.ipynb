softmax with one sample set, (vector)

```python
import numpy as np
```

```python
a = np.random.randn(5)
```

```python
a
```

```python
expa = np.exp(a)
```

```python
expa
```

```python
answer = expa/ expa.sum()
```

```python
answer
```

```python
answer.sum()
```

softmax with multiple sample set (matrix)

```python
A = np.random.randn(100, 5)
```

```python
expA = np.exp(A)
```

```python
expA
```

```python
answer = expA / expA.sum(axis=1, keepdims=True)
```

```python
answer.sum(axis=1)
```
