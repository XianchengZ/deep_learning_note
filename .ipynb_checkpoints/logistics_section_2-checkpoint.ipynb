## Section 2: Basics

```python
import numpy as np
```

```python
N = 100
```

```python
D = 2
```

```python
X = np.random.randn(N,D)
```

```python
ones = np.array([[1]]*N)
```

```python
Xb = np.concatenate((ones, X), axis=1)
```

```python
w = np.random.randn(D + 1)
```

```python
w
```

```python
z = Xb.dot(w)
z
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
sigmoid(z)
```
